```python
# Create religion-based variation dictionary from the province_city_weights data
religion_based_variation = {}

for province, cities in province_city_weights.items():
    # Calculate average religious distribution for the province
    religion_totals = {
        "buddhism": 0,
        "hinduism": 0,
        "islam": 0,
        "roman_catholic": 0,
        "other_christian": 0,
        "other": 0
    }

    total_population_weight = 0

    for city, data in cities.items():
        pop_weight = data["population_from_province"]
        total_population_weight += pop_weight

        for religion, percentage in data["religions"].items():
            religion_totals[religion] += percentage * pop_weight

    # Calculate weighted average percentages for the province
    province_religion_distribution = {}
    for religion, total in religion_totals.items():
        province_religion_distribution[religion] = total / total_population_weight

    religion_based_variation[province] = province_religion_distribution


def weighted_choice(weight_dict):
    """Pick a key based on weight distribution"""
    total = sum(weight_dict.values())
    r = random.uniform(0, total)
    upto = 0
    for k, w in weight_dict.items():
        if upto + w >= r:
            return k
        upto += w
    return k


def generate_customers(num_customers):
    customers = []
    cust_id = 1

    for province, info in SriLanka_Fish_Ecommerce_Interest.items():
        expected_customers = round(num_customers * info["interest_score"])

        for _ in range(expected_customers):
            # Select religion based on province's religious distribution
            religion = weighted_choice(religion_based_variation[province])

            # Select age group based on buying interest
            age_group = weighted_choice(buying_interest)

            # Create city weight dictionary for this province (extract population weights only)
            city_weights = {}
            for city, data in province_city_weights[province].items():
                city_weights[city] = data["population_from_province"]

            # Select city based on population weights
            city = weighted_choice(city_weights)

            customers.append({
                "CustID": cust_id,
                "Age Group": age_group,
                "City": city,
                "Province": province,
                "Religion": religion
            })
            cust_id += 1

    return customers


# --- Generate dataset ---
customers = generate_customers(2304)

# --- Save full dataset ---
with open("customer_dataset.json", "w", encoding="utf-8") as f:
    json.dump(customers, f, indent=2, ensure_ascii=False)

print("âœ… Dataset saved as customer_dataset.json")
print(f"ðŸ“Š Generated {len(customers)} customers")

# Display statistics
print("\nðŸ“ˆ Customer Distribution by Province:")
province_counts = {}
for customer in customers:
    province = customer["Province"]
    province_counts[province] = province_counts.get(province, 0) + 1

for province, count in sorted(province_counts.items(), key=lambda x: x[1], reverse=True):
    percentage = (count / len(customers)) * 100
    print(f"  {province}: {count} customers ({percentage:.1f}%)")
```
