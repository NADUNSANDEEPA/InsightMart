```python
def random_date(start_date, end_date):
    """Generate a random datetime between two dates."""
    delta = end_date - start_date
    random_days = random.randint(0, delta.days)
    return start_date + timedelta(days=random_days)


def get_random_buy_date():
    """Generate random buy date (Jan 1 - Dec 31, 2024)."""
    start = datetime(2024, 1, 1)
    end = datetime(2024, 12, 31)
    return random_date(start, end)

def check_festival_products(date):
    """Return list of boosted product codes if date falls in a festival period."""
    boosted = []
    for fest, details in seasonal_and_festival_periods.items():
        if details['start'] <= date <= details['end']:
            boosted.extend(details['products'])
    return boosted
```

```python
def generate_record(customer, customer_id):
    records = []

    # Generate a random buy date
    buy_date = get_random_buy_date()
    boosted_products = check_festival_products(buy_date)

    # Family size
    family_size = random.randint(1, 7)
    consumption = seafood_and_meat_daily_consumption_by_family_size[family_size]

    # Religion preferences
    religion = customer.get("Religion", "Others").lower()
    prefs = religios_based_meat_preferences.get(religion, religios_based_meat_preferences["Others"])

    fish_or_meat_random = random.randint(0, 1)
    possible_products = products.copy()

    # Daily consumption
    if fish_or_meat_random == 0:
        fish_items = int(consumption['fresh_fish_items_per_day'] * consumption["fresh_fish_items_per_day"])

        # Fish purchases
        for _ in range(fish_items):
            fish_products = [p for p in possible_products if "Fish" in p['category'] or "Seafood" in p['category']]
            if not fish_products:
                continue
            product = random.choice(fish_products)
            quantity = random.randint(1, 3)

            record = {
                'Customer Id': customer_id,
                'Customer Age Group': customer.get("Age Group", "Unknown"),
                'City': customer.get("City", "Unknown"),
                'Province': customer.get("Province", "Unknown"),
                'Product Code': product['code'],
                'Product Name': product['name'],
                'Product Category': product['category'],
                'Quantity': quantity,
                'Unit Price': product['unit_price'],
                'Discount': f"{random.uniform(0, 0.15):.2%}",
                'Buying Date': buy_date.strftime('%Y-%m-%d'),
                'Payment Method': random.choice(payment_methods),
                'Order Status': random.choice(order_status_options),
                'Delivery Type': random.choice(delivery_types),
                'Delivery Status': random.choice(delivery_status_options),
                'Rate': random.randint(1, 5)
            }
        records.append(record)

    else:   
        meat_items = int(consumption['meat_items_per_day'] * consumption["meat_items_per_day"])
        
        # Meat purchases
        for _ in range(meat_items):
            # Select product based on religion preferences
            meat_products = [p for p in possible_products if "Meat" in p['category'] or "Pork" in p['category']]
            if not meat_products:
                continue
            product = random.choice(meat_products)

            # Religious filtering
            if "beef" in product['name'].lower() and prefs.get("beef") in ["very_low", "low"]:
                continue
            if "pork" in product['name'].lower() and prefs.get("pork") in ["very_low", "low"]:
                continue

            quantity = random.randint(1, 3)

            record = {
                'Customer Id': customer_id,
                'Customer Age Group': customer.get("Age Group", "Unknown"),
                'City': customer.get("City", "Unknown"),
                'Province': customer.get("Province", "Unknown"),
                'Product Code': product['code'],
                'Product Name': product['name'],
                'Product Category': product['category'],
                'Quantity': quantity,
                'Unit Price': product['unit_price'],
                'Discount': f"{random.uniform(0, 0.15):.2%}",
                'Buying Date': buy_date.strftime('%Y-%m-%d'),
                'Payment Method': random.choice(payment_methods),
                'Order Status': random.choice(order_status_options),
                'Delivery Type': random.choice(delivery_types),
                'Delivery Status': random.choice(delivery_status_options),
                'Rate': random.randint(1, 5)
            }
            records.append(record)

    # Boosted seasonal products
    for boosted_code in boosted_products:
        product = next((p for p in products if p['code'] == boosted_code), None)
        if product:
            record = {
                'Customer Id': customer_id,
                'Customer Age Group': customer.get("Age Group", "Unknown"),
                'City': customer.get("City", "Unknown"),
                'Province': customer.get("Province", "Unknown"),
                'Product Code': product['code'],
                'Product Name': product['name'],
                'Product Category': product['category'],
                'Quantity': random.randint(2, 5),
                'Unit Price': product['unit_price'],
                'Discount': f"{random.uniform(0, 0.20):.2%}",
                'Buying Date': buy_date.strftime('%Y-%m-%d'),
                'Payment Method': random.choice(payment_methods),
                'Order Status': random.choice(order_status_options),
                'Delivery Type': random.choice(delivery_types),
                'Delivery Status': random.choice(delivery_status_options),
                'Rate': random.randint(3, 5)
            }
            records.append(record)

    return records


# === Generate full dataset ===
dataset = []
for idx, customer in enumerate(customers, start=1):
    dataset.extend(generate_record(customer, idx))

# Save dataset
with open("generated_sales_dataset.json", "w", encoding="utf-8") as f:
    json.dump(dataset, f, indent=4, ensure_ascii=False)

print(f"✅ Generated {len(dataset)} purchase records saved to generated_sales_dataset.json")
```
